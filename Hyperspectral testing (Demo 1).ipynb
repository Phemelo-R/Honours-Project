```python
import matplotlib.pyplot as plt
from spectral import open_image, imshow
from pyproj import Transformer
```

```python
img = open_image('ang20231022t092801_000_L1B_ORT_main_46dd9a4a_LOC.hdr')
print(img)
```

```python
imshow(img)
```

```python
# Analyse Pixels
pixel_spectrum = img[100, 150]  
print(pixel_spectrum)
```

```python
plt.plot(pixel_spectrum)
plt.title("Spectral Profile for Pixel (100, 150)")
plt.xlabel("Band Number")
plt.ylabel("Reflectance")
plt.show()
```

```python
from pyproj import Transformer

# Change CRS
transformer = Transformer.from_crs("EPSG:4326", "EPSG:32633", always_xy=True)

# Plot Area
lon, lat = 25.0, -26.0  
easting, northing = transformer.transform(lon, lat)
print(f"Easting: {easting}, Northing: {northing}")
```
